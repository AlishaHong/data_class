## 실습에 사용할 예제 파일

[국내 아이돌 평판지수 (csv)](http://bit.ly/ds-korean-idol)

```
import pandas as pd
```

## 1. 파일 읽어오기 (csv)

```
# csv파일이 구글드라이브에 저장된 주소의 '약식 주소'
df = pd.read_csv('http://bit.ly/ds-korean-idol')
```

```
df
```

```
# head() : 맨 앞의 5행의 데이터를 출력한다.
df.head()
```

```
# tail() : 맨 끝에서5개의 데이터를 출력한다.
df.tail()
```

### 2-1. column(열) 출력하기

```
# pandas에서 dtype이 object인 경우 -> 문자열
df.columns
```

### 2-2. index(행) 출력하기

```
df.index
```

### 2-3. info 는 기본적인 row(행)의 정보와 데이터 타입을 알려줍니다.

**tip.** info 메소드는 주로 빠진 값 (null 값)과 데이터 타입을 볼 때 활용합니다.

```
# column의 갯수, 컬럼별 Not Null count의 갯수, 컬럼별 데이터 타입
df.info()
```

## 7. 정렬하기 

### 7-1. 오름차순 정렬 (default)

```
# sort_index()함수는 index를 기준으로 정렬해주는 함수
df.sort_index()
```

### 7-2. 내림차순 index 정렬

```
df.sort_index(ascending=False)
```

## 8. column 별로 정렬

### 8-1. 오름차순 정렬

```
df.sort_values(by='키')
```

### 8-2. 내림차순 정렬

```
df.sort_values(by='키', ascending=False)
```

### [BONUS] 복수 정렬

```
# 정렬 기준 1번 :'키', 키가 동일한 경우 정렬기준 2번
df.sort_values(by=['키', '브랜드평판지수'])
```

```
df.sort_values(by=['키', '브랜드평판지수'], ascending=False)
```
