```python
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

#플랫폼 패키지는 현재 내 컴퓨터의 os가 무엇인지 코드로 확인하는 패키지
#맥과 윈도우에서 다 사용할 수 있는 코드
#윈도우용 코드만 가져다 써도됨
if platform.system() == 'Darwin':   #macOS
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':   #WINDOWS
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:   #LINUX
    print('Unknown system... sorry~~~~') 
```

#### 폰트설정하기

```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
# plt.xlabel('X-Axis', labelpad=15, fontdict=font1)
# plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)

# 폰트관련 매개변수 잘 봐두기!
plt.xlabel('X-Axis', labelpad=15, fontsize = 14, fontfamily = 'serif', fontweight = 'bold', color =  'blue')
plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
```

```python
# 축 레이블 위치 정하기
# x축은 left,center,right
# y축은 top,center,bottom


plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=20, fontdict = font1, loc = 'right')
plt.ylabel('Y-Axis', labelpad=20, fontdict = font2, loc = 'bottom')
```

#### 선 종류 지정하기

```python
# plt.plot([1,2,3],[4,4,4], '-', color = 'r', label = 'solid')
# linestyle에 지정해줘도 되고 직접 써도 된다.
# 포맷 문자열과 같이 ‘solid’, ‘dashed’, ‘dotted’, dashdot’ 
# 네가지의 선 종류를 지정할 수 있습니다.
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='r', label="'solid'")
plt.plot([1,2,3],[3,3,3], '--', color = 'r', label = 'Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc = 'upper left', ncol = 4)
```

#### 마커/색상

```python
plt.plot([1, 2, 3], [4, 4, 4], 'o:', color = 'lightcoral', label='solid')
#마커는 세번째로
plt.plot([1,2,3],[3,3,3], '--', color = 'violet', label = 'Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='olive', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='rosybrown', label='Dash-dot')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc = 'upper left', ncol = 4)
```

## 그리드 설정

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, '-o', color = 'gold')
plt.plot(x, x**2, color='lightblue', marker='*', linewidth=2)
plt.plot(x, x**3, color='lightpink', marker='^', markersize=9)
plt.grid()
# plt.grid(axis = 'x')
plt.grid(axis = 'y')
#한글폰트 설치는 파일 윗부분에서 먼저 진행함
plt.title('그래프 연습', fontfamily = 'malgun Gothic', fontweight = 'bold',loc = 'right', pad = 20)
plt.show()
```
