```python
#[질문1] : 평창군 전체에 한해동안 가장 방문객이 많았던 해는 언제인가?

#1월부터 12월까지의 방문객을 모두 더한 '총방문객수'컬럼 만들기
df['총방문객수'] = df.loc[:,'1월':'12월'].sum(axis =1 )
print(df['총방문객수'])

#연도별로 총 방문객수의 총합
year_visitor_sum = df.groupby('연도')['총방문객수'].sum()
print(year_visitor_sum)

#가장 많은 방문객 수 총합
result = year_visitor_sum.max()
print(result)

#result가 가장 높은 행의 인덱스 출력
max_year = year_visitor_sum.idxmax()
print(max_year)
```

```python
#1번문제 강사님 코드

def yearTotal(year):
    # 특정 해의 데이터만 선택해서 데이터프레임 만들기
    df_year = df.loc[df['연도'] == year] 

    # 관광지별 모든 1~12월 방문객수를 더한 총방문객수 계산
    df_year['총방문객수'] = df_year.loc[:,'1월':'12월'].sum(axis=1)

    # 2010년의 관광지별 방문객수 모두 더하기
    total = df_year['총방문객수'].sum()  # 7183542.0
    return total

for year in range(2010, 2022):
    print(f'{year}년 : {yearTotal(year)}')
```

```python
#[질문2] : 평창군 전체에 어떤달에 가장 방문객이 많았던가?

#월별 합산 구해서 
#최대값구하면됨(정확히는 최대값을 가진 인덱스구하기)
#시리즈를 반환함
#pandas가 데이터프레임의 열을 기준으로 합계를 계산한 결과를 저장할 때, 
#열 이름을 인덱스로 하고 합계 값을 값으로 하는 Series 객체를 반환하는 것이 일반적

monthly_sums1 = df.loc[:,'1월':'12월'].sum(axis = 0)
print(monthly_sums1)
print(type(monthly_sums1))
print(monthly_sums1.idxmax())



months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
monthly_sums2 = df[months].apply(lambda x: x.sum(skipna=True))
print(monthly_sums2)



monthly_sums3 = df.loc[:,'1월':'12월'].apply(lambda x: x.sum(skipna=True))
print(monthly_sums3)
```

```python
#2번 강사님 코드

# 2. 결측치 있는 행 삭제 (4~12월 데이터가 없는 행 삭제)
df.dropna(how='any', inplace=True)

def yearTotal(year):
    # 특정 해의 데이터만 선택해서 데이터프레임 만들기
    # 물리적으로 다른 데이터 저장공간 df_year를 생성해서 아래 코드를 처리
    df_year = df.loc[df['연도'] == year].copy() 

    # 관광지별 모든 1~12월 방문객수를 더한 총방문객수 계산
    df_year['총방문객수'] = df_year.loc[:,'1월':'12월'].sum(axis=1)

    # 2010년의 관광지별 방문객수 모두 더하기
    total = int(df_year['총방문객수'].sum())  # 7183542.0
    return total

for year in range(2010, 2022):
    print(f'{year}년 : {yearTotal(year)}')
```

```python
#[질문3] : 평창군에서 가장 핫 플레이스는? (최근 5년동안) 5년 평균 방문객이 가장 높은 곳
# --> 정답은 오대산 국립공원

#최근5년동안의 데이터만 따로 뽑아보자
#총방문객컬럼이 이미 있으므로 5년의 값을 모두 더해 평균내기 


#1. 무조건 더해서 1등구하기!!

#대략 2018년을 기준으로 잡았지만 
#연도중 가장 큰 값을 구해 그값에서 5를 뺀 값을 2018대신 넣어도 좋다
recent_year = df.loc[df['연도']>=2018]
print(recent_year)

#관광지별로 2018년 이후의 모든 총방문객수를 더한 값을 표현한 데이터
#전체 데이터 말고 2018년 이후만 따로 모아놓은 데이터(recent_year)를 기준으로 구한다.
place_sum = recent_year.groupby('관광지')['총방문객수'].sum() #관광지 별 연도별 총방문객수의 합

#가장 큰 값을 가진 데이터의 인덱스(관광지)
#idxmax 대박 유용함
print(f'총합 가장 높은곳 : {place_sum.idxmax()}')



#2. 연도의 갯수를 구해서 평균내기
#해당 관광지가 총 방문객수라는 정보를 보유한 연도가 몇개인지 구하자 
#최근 5년이라고 했으니 그냥 5로 나눌수도 있지만 위의 방법이 정확하겠지?
year_count = recent_year.groupby('관광지')['총방문객수'].count()
place_sum = recent_year.groupby('관광지')['총방문객수'].sum()

#평균
avg_visitors = place_sum/year_count
#그냥 5로 나누어 구한 평균
five = place_sum/5


# print(avg_visitors)
print(f'평균 가장 높은곳 : {avg_visitors.idxmax()}')
print(f'5로 나눈 평균이 가장 높은곳 : {five.idxmax()}')


# 3가지 방법 모두 결과값은 오대산 국립공원! 
```

```python
#[질문4] : 2010년 ~ 2022년에 이르기까지 관광지가 추가되거나, 삭제된 곳은 어디인가?

#2010~2022까지 쭉 지속된 관광지가 아니라면 저 질문의 답에 해당되지 않을까..
#2011년부터 데이터가 있다면 중간에 추가된것이고
#2022년 안에 데이터가 더이상 추가되지 않는다면 삭제된 것이니까.. 



# 2010년부터 2022년까지 연도 필터링
year_range = range(2010, 2023)
df_filtered = df[df['연도'].isin(year_range)]

# 관광지별 연도 범위 확인
first_year = df_filtered.groupby('관광지')['연도'].min()
last_year = df_filtered.groupby('관광지')['연도'].max()

# 2010년 이후에 추가된 관광지
added_sites = first_year[first_year > 2010].index.tolist()

# 2022년 전에 마지막으로 등장한 관광지
deleted_sites = last_year[last_year < 2022].index.tolist()

# 2010년부터 2022년까지 계속 있던 관광지 
# full_sites_first = first_year[first_year == 2010]
# full_sites_last = last_year[last_year == 2022]

# all = pd.concat([full_sites_first, full_sites_last])
# duplicated = all[all.index.duplicated(keep=False)]


print(f"추가된 관광지: {added_sites}")
print(f"삭제된 관광지: {deleted_sites}")
```
