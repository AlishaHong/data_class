```python
# y = wx + b
# 조금더 복잡하게 다항회귀를 해볼수도 있다.

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# 데이터 준비
np.random.seed(14)
n = 100 
x = np.arange(1,n+1)    #1차원

#난수를 더해 데이터에 노이즈를 준다.
randvalue = np.random.random(n)*35
randvalue = randvalue.reshape(-1,1)
x = x.reshape(-1,1)     #2차원으로 변환
y = 4*x + 7     #x로 인해 브로드캐스트연산 일어남 
y = y + randvalue

y
```

```python
print(x.shape) #10행 1열
print(y.shape)
```

```python
# 모델링

le = LinearRegression()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y,color='pink')
```

```python
#훈련데이터 테스트데이터 분리하기 

from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
le.fit(train_x,train_y) #모델학습
```

```python
# pred_y = le.predict([[20],[21]])
# pred_y

pred_y = le.predict(test_x)
pred_y
```

```python
le.score(test_x,test_y) 
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(mean_squared_error(test_y, pred_y))
print(mean_absolute_error(test_y, pred_y))
print(np.sqrt(mean_squared_error(test_y, pred_y)))
```

```python
# 다중선형모델


# y = w1*x1 + w2*x2 + b
#w1 = 4 , x2 = 5 , b=3
# y = 4*x1 + 5*x2
np.random.seed(1)
n = 100
x1 = np.arange(n).reshape(-1,1)
x2 = np.arange(n).reshape(-1,1)

x3 = pd.DataFrame(np.hstack([x1,x2]), columns=['x1','x2'])
print(x3)
randvalue = np.random.random(n)*50
randvalue = randvalue.reshape(-1,1)
y = 4*x1 + 5*x2 + 3

# print(x1)
# print(x2)
# print(y)

y = randvalue + y 
print(y)
```

```python
# x3 데이터프레임과 y로 훈련/테스트 세트 분리

train_x,test_x,train_y,test_y = train_test_split(x3,y, test_size=0.2,random_state=1)
print(train_x)
print(type(train_x))
le2 = LinearRegression()    #모델생성

le2.fit(train_x,train_y)    #학습데이터 훈련
pred_y = le2.predict(test_x)
le2.score(test_x,test_y)
```

```python
# 수식에 3제곱을 구현하고 모델링하기
# 다항회귀

from sklearn.preprocessing import PolynomialFeatures


np.random.seed(14)
n = 100

x = np.arange(n)
x = x.reshape(-1,1)

randvalue = np.random.random(n) * 350000
randvalue = randvalue.reshape(-1,1)


y = 2 * x**3  + x + 2 + randvalue

poly_features = PolynomialFeatures(degree=3)
x_poly = poly_features.fit_transform(x)


train_x, test_x, train_y,test_y = train_test_split(x_poly ,y, test_size=0.2, random_state=1)

le = LinearRegression()
le.fit(train_x,train_y)
pred_y = le.predict(test_x)

print(test_x)

plt.scatter(test_x[:,1], test_y)
# test_x에서 원래 x의 값이 1번째컬럼이므로 1로 검색
plt.scatter(test_x[:,1], pred_y, color= 'red')
```

```python
n = 100

randNum = np.random.rand(n,1)
x = 6*randNum - 3

np.random.seed(14)
randValue =np.random.random(n)*20
randValue = randValue.reshape(-1,1)

y = 2 * x**3 + x + 2 + randValue
plt.scatter(x,y)
plt.show()
```

```python
# 고차방정식 함수 

from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=4, include_bias=True)
x_poly = poly_features.fit_transform(x)
x[0], x_poly[0]
```

```python
model = LinearRegression()
model.fit(x_poly,y)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
#파이프라인

from sklearn.pipeline import make_pipeline
model = make_pipeline(PolynomialFeatures(degree=3, include_bias=True), LinearRegression())
model.fit(x,y)
```
