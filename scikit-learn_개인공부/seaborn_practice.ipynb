```python
# 한글 폰트 설치와 설정

import platform
from matplotlib import font_manager, rc

#플랫폼 패키지는 현재 내 컴퓨터의 os가 무엇인지 코드로 확인하는 패키지
#맥과 윈도우에서 다 사용할 수 있는 코드
#윈도우용 코드만 가져다 써도됨
if platform.system() == 'Darwin':   #macOS
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':   #WINDOWS
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:   #LINUX
    print('Unknown system... sorry~~~~') 
    

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic') 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 불러오기 (예시: 붓꽃 데이터셋)
iris = sns.load_dataset('iris')

# 1. 히스토그램 & 커널 밀도 추정(KDE) 곡선 
sns.distplot(iris['sepal_length'], kde=True, rug=True)
plt.title('Sepal Length 분포')
plt.show()

# 2. 박스플롯
sns.boxplot(x='species', y='sepal_width', data=iris)
plt.title('Sepal Width 분포 (종별)')
plt.show()

# 3. 바이올린 플롯
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.title('Sepal Width 분포 (종별)')
plt.show()

# 4. 스웜 플롯
sns.swarmplot(x='species', y='sepal_width', data=iris)
plt.title('Sepal Width 분포 (종별)')
plt.show()

# 5. 2차원 분포 시각화: Jointplot
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='kde')
plt.show()
```
