```python
#geopy 패키지
#주소값을 위경도로 변환

# 가입 없이 주소->좌표 변환
from geopy.geocoders import Nominatim

# 주소나 위도, 경도가 중복되지 않은 고유한 정보를
# 리스트에 각각 담아주기 위해 빈 리스트 생성

unique_lat=[]
unique_lng=[]
# except_addr=[]


#지오코딩
#주소를 마커나 지도를 배치하는데 사용할 수 있는 지리좌표로 변환하는 과정

#역지오코딩 
#지리좌표를 사람이 읽을 수 있는 주소로 변환 

# 주소를 위도와 경도로 변환해 주는 함수
def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)    #사용자 에이전트 설정
    geo = geolocoder.geocode(address)   #주소로 위치정보 가져오기(위도와 경도 반환)
    # 주소값을 제대로 인식하지 못할때
    # 위도와 경도값을 None값을 리턴한다.
    if geo==None:
        print(f'주소가 어디인지 인식할 수 없습니다.: {address}')
        except_addr.append(address)
        return 0, 0
    else:
        return geo.latitude, geo.longitude  #위도,경도
```

```python
#컬럼을 더 만들어서 위도와 경도를 추가해준다. 
unique_lat=[]
unique_lng=[]

map_addr = df['설치장소'].unique()
print(map_addr)

for addr in map_addr:
    if addr=="제주특별자치도 제주시 삼도이동":
        addr = "제주시 삼도 2동"
        
    geo = geocoding(addr)
    print(geo)
    unique_lat.append(geo[0])
    unique_lng.append(geo[1])
    
```

```python
len(df['설치장소'].unique()), len(unique_lat), len(unique_lng)
```

```python
map_df = pd.DataFrame({'설치장소':df['설치장소'].unique(),
                       '위도':unique_lat,
                       '경도':unique_lng})

map_df
```

```python
map_df.set_index('설치장소',inplace=True)
```

```python
# 지도를 그릴때 사용하는 패키지는 folium

import pandas as pd
import folium

# 지도가 그려질 중심 좌표(대략 제주도의 중심 좌표), 지도를 그릴때 얼마나 줌을 적용할지
jeju_map = folium.Map(location=[33.38, 126.55], zoom_start=11)

for i in range(len(map_df)):
    pos_list = [map_df.iloc[i][0], map_df.iloc[i][1]]
    folium.Marker(pos_list, popup=map_df.index[i]).add_to(jeju_map)
    
jeju_map.save('jeju.html')
```
