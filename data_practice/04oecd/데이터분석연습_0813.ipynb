```python
# 데이터읽어오기 
import pandas as pd

oecd = pd.read_excel('oecd.xlsx', header = 2)
oecd.head()
oecd.tail()
```

```python
#데이터전처리하기 

#1. 결측치 삭제해버리기 
#2. 인덱스명 수정 
#3. 행과 열 전치 
```

```python
# 인덱스명 수정 

oecd.rename(columns={'Unnamed: 0':'index'}, inplace=True)
oecd

oecd.set_index('index',inplace=True,drop=True)

oecd = oecd.loc['영국':'아이슬란드']
```

```python
oecd
```

```python
oecd = oecd.T
oecd
```

```python
oecd
```

```python
# 문자열 -> 정수형으로 변환하기 

oecd = oecd.replace(',','',regex=True).astype(int)
oecd
```

```python
import matplotlib.pyplot as plt

#함수만들기 
#x축은 연도, y축은 전력사용 평균값

def graph(data,countries):
    x = oecd.index
    
    plt.figure(figsize=(12,8))
    for country in countries:
        
        plt.plot(x, data[country], label= country, marker = 'o')
        
    
    plt.xlabel('연도별')
    plt.ylabel('전력사용량(GWh)')
    plt.legend()
    plt.grid()
    
```

```python
graph(oecd,['한국','미국','영국'])
```
